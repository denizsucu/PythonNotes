```python
import numpy as np
import matplotlib.pyplot as plt
```

*diğer editörlerde geçerli değil, grafikleri kolayca göstermeye olanak tanır (%matplotlib inline)

```python
%matplotlib inline
```

```python
ageList = [10, 20, 30, 30, 30, 40, 50, 60, 70, 80]
```

```python
weightList = [25, 40, 60, 65, 70, 80, 85, 90, 85, 80]
```

```python
npAgeArray = np.array(ageList)
npWeightArray = np.array(weightList)
```

```python
npAgeArray
```

```python
npWeightArray
```

```python
plt.plot(npAgeArray, npWeightArray, "g") #ilk yazılan x eksenini belirliyor, ikinci y, "g" => renk
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Kilonun Yaşa göre Değişimi")
#plt.show() # burada şimdilik gerek yok inline dedik
```

```python
numpyArray1 = np.linspace(0, 10, 20)
```

```python
numpyArray1
```

```python
numpyArray2 = numpyArray1**2
```

```python
numpyArray2
```

```python
plt.plot(numpyArray1, numpyArray2, "b--") #kesik çizgi halinde olur
```

```python
plt.plot(numpyArray1, numpyArray2, "r*-") # noktaları yıldızla işaretler
```

```python
plt.subplot(1, 2, 1) # Bir satır, 2 sütun olacak. 1. grafiği çiziyorum demek
plt.plot(numpyArray1, numpyArray2, "g*-")
plt.subplot(1, 2, 2) # 2. grafiği çiziyoruz
plt.plot(numpyArray2, numpyArray1, "r*-")
```

```python
myFigure = plt.figure() # Boş bir canvas gibi
figureAxes = myFigure.add_axes([0.2, 0.2, 0.9, 0.9]) # axes eklemek, [left sp, bottom sp, width, height]
figureAxes.plot(numpyArray1, numpyArray2, "b")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
# Bu işlemleri aynı cell içinde yapmak gerekiyor
```

```python
figure2 = plt.figure()

graph1 = figure2.add_axes([0.1, 0.1, 0.9, 0.9])
graph2 = figure2.add_axes([0.2, 0.4, 0.3, 0.3]) # 1.2 gibi bir değer versem sağda çizilecek

graph1.plot(numpyArray1, numpyArray2, "r")
graph1.set_xlabel("X Ekseni")
graph1.set_ylabel("Y Ekseni")
graph1.set_title(" Ana Grafik Başlık")

graph2.plot(numpyArray2, numpyArray1, "y")
graph2.set_xlabel("X Ekseni")
graph2.set_ylabel("Y Ekseni")
graph2.set_title("Küçük Grafik Başlık")
```

```python
(figureX, axesX) = plt.subplots() # Tuple döner
# figureX # Boş bir figure
# axesX # Boş bir eksen
axesX.plot(numpyArray1, numpyArray2, "r") # add_axes falan demeye gerek kalmadı
```

```python
(figure3, axeses) = plt.subplots(nrows = 1, ncols=2)
for eksen in axeses:
    eksen.plot(numpyArray1, numpyArray2, "m")
    eksen.set_xlabel("X Ekseni")
    eksen.set_ylabel("Y Ekseni")
plt.tight_layout() # gmrüntüyü daha düzgün hale getiriyor (arayı falan açtı)
# figure3 # burada çizildi ama çizilmeme durumunda figure'u çağırmak lazım
```

```python
newFigure = plt.figure() # figure(figsize=(3,3)) => bu şekilde de size verebiliriz
# figure(dpi=150) => çözünürlüğünü ve büyüklüğünü ayarlayabiliriz => kaydederken kullanmak daha mantıklı

newAxes = newFigure.add_axes([0.1, 0.1, 0.9, 0.9])

newAxes.plot(numpyArray1, numpyArray1 ** 3, label = "numpyArray ** 3")
newAxes.plot(numpyArray1, (numpyArray1 ** 3)/4, label = "(numpyArray ** 3)/4")
newAxes.legend(loc=2) # yer tanımlayabiliriz
```

```python
newFigure.savefig("myFigure.png", dpi=200) # grafiği kaydeder
```
