```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyArray1 = np.linspace(0, 10, 20)
```

```python
numpyArray1
```

```python
numpyArray2 = numpyArray1 ** 2
```

```python
numpyArray2
```

```python
(myFigure, myAxes) = plt.subplots()
myAxes.plot(numpyArray1, numpyArray2, color = "#3A89AF", alpha = 0.3) # alpga => transparency
myAxes.plot(numpyArray2, numpyArray1, color = "aqua")
```

```python
(newFigure, newAxes) = plt.subplots()
newAxes.plot(numpyArray1, numpyArray1 + 2, color = "magenta", linewidth=4.0)
newAxes.plot(numpyArray1, numpyArray1 + 3, color = "aqua", linewidth= 2.0)

newAxes.plot(numpyArray1, numpyArray1 + 4, color = "brown", linestyle="--")
newAxes.plot(numpyArray1, numpyArray1 + 5, color = "orange", linestyle="-.")
newAxes.plot(numpyArray1, numpyArray1 + 6, color = "blue", linestyle=":")

newAxes.plot(numpyArray1, numpyArray1 + 7, color = "black", linestyle="--", marker = "*",markersize=10.0, markerfacecolor="red")
newAxes.plot(numpyArray1, numpyArray1 + 8, color = "purple", linestyle="-", marker = "+",markersize=7, markerfacecolor="black")
```
